```python
IP = df_c.drop('label', axis=1)

OP = df_c['label']
```

```python
OP[OP == "DDoS"] = 0
OP[OP == "C&C"] = 1
OP[OP == "Benign"] = 2
OP[OP == "PartOfAHorizontalPortScan"] = 3
```

```python
OP.unique()
```

__conversion of the dataset to an optimized data structure called Dmatrix__

```python
for column in IP.columns:
    if IP[column].dtype == 'object':
        IP[column] = IP[column].astype('category')
    else:
        IP[column] = pd.to_numeric(IP[column])

OP = OP.astype('int')
```

```python
OP.unique()
```

```python
try:
    data_dmatrix = xgb.DMatrix(data=IP, label=OP, enable_categorical=True)
    print("DMatrix created successfully")
except ValueError as e:
    print(f"Error creating DMatrix: {e}")
```

```python
data_dmatrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(IP, OP, test_size = 0.3, random_state = 0)
```

```python
X_train.shape[0], X_train.shape[1]
```

```python
params = {
    'objective': 'multi:softmax', 
    'max_depth': 4,
    'alpha': 10,
    'learning_rate': 1.0,
    'n_estimators': 100,
    'num_class': 4
}
            
xgb_clf = XGBClassifier(**params)
xgb_clf.fit(X_train, y_train)
```

```python
y_pred = xgb_clf.predict(X_test)
```

```python
print('XGBoost model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
params = {
    'objective': 'multi:softmax', 
    'num_class': 4,               
    'colsample_bytree': 0.3,       
    'learning_rate': 0.1,          
    'max_depth': 5,                
    'alpha': 10
}

xgb_cv = cv(
    dtrain=data_dmatrix,
    params=params,
    nfold=5,
    num_boost_round=50,
    early_stopping_rounds=10,
    metrics="mlogloss",
    as_pandas=True,
    seed=123
)
```

```python
best_num_boost_round = xgb_cv.shape[0]
bst = xgb.train(params, data_dmatrix, num_boost_round=best_num_boost_round)
```

```python
pipe = Pipeline([("scaler", StandardScaler()), ("xgb", XGBClassifier(n_estimators=best_num_boost_round, 
                                                                      max_depth=5, 
                                                                      learning_rate=0.1, 
                                                                      colsample_bytree=0.3, 
                                                                      alpha=10, 
                                                                      objective='multi:softmax', 
                                                                      num_class=4))])

pipe.fit(X_train, y_train)
```

```python
update_registered_converter(
    XGBClassifier,
    "XGBoostXGBClassifier",
    calculate_linear_classifier_output_shapes,
    convert_xgboost,
    options={"nocl": [True, False], "zipmap": [True, False, "columns"]},
)
```

```python
try:
    model_onnx = convert_sklearn(
        pipe,
        "pipeline_xgboost",
        [("input", FloatTensorType([None, X_train.shape[1]]))],
        target_opset={"": 12, "ai.onnx.ml": 2},
    )

    with open("pipeline_xgboost.onnx", "wb") as f:
        f.write(model_onnx.SerializeToString())

    print("Model converted and saved successfully.")

except Exception as e:
    print(f"An error occurred during the conversion: {e}")
```
